```python

#BIBLIOTECAS
import datetime #Manipulação de datas
import pandas as pd #Armazena a estrutura de dados dataframe
import pandas as pd_ #Armazena a estrutura de dados dataframe
import matplotlib.pyplot as plt #Geração de gráficos
import calendar #Manipulação de datas

# Widgets imports:
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets #Ambiente gráfico Menu

def calculadora_rf(cotacao_inicial, juro_anual, dia_inicio, mes_inicio, ano_inicio, dia_fim, mes_fim, ano_fim):

    # Declaração da lista que vai armazenar o cálculo dos juros, inicialmente
    conf_ = {'Data': [],
             'Cotação': []}

    # Conversão de datas para formato datetime
    data_inicial = datetime.datetime.strptime(str(ano_inicio) + "-" + str(mes_inicio) + "-" + \
                                              str(dia_inicio) + " 00:00:00", "%Y-%m-%d %H:%M:%S")

    data_final = datetime.datetime.strptime(str(ano_fim) + "-" + str(mes_fim) + "-" + \
                                            str(dia_fim) + " 00:00:00", "%Y-%m-%d %H:%M:%S")

    # Conversão da cotacao_atual para formato float
    cotacao_atual = float(cotacao_inicial)

    # O Python soma em segundos
    # Um dia tem 86400 segundos
    one_day = 86400

    # Um ano tem one_day * 365 segundos, a menos que seja bissexto
    one_year = one_day * 365
    one_leap_year = one_day * 366

    # Não é possível somar datetime.datetime.strptime com o inteiro 86400
    # Por isso é necessário criar um inteiro e um .timestamp(), pois
    # ambos podem ser somados
    data_aux = data_inicial.timestamp()

    # É necessário manter uma variável data no formato string
    # pois a função isleap utiliza o ano em formato string
    # para determinar se ele é bissexto ou não
    data_str = str(datetime.datetime.fromtimestamp(data_aux))

    # Datas do mesmo tipo datetime.datetime.strptime podem ser comparadas
    data_contador = data_inicial
    while (data_contador < data_final):

        # Variável ano será utilizada pela função isleap para determinar
        # se o ano é bissexto
        ano = data_str[0:4]

        # A cada iteração, adiciona na lista a data e a cotação
        # acrescida dos juros
        try:
            conf_['Cotação'].append(float(cotacao_atual))
            conf_['Data'].append(data_contador)
        except IndexError as error:
            pass

        # Prepara as variáveis de data para cálculo dos juros do próximo ano
        if(calendar.isleap(int(ano))):
            data_contador = datetime.datetime.fromtimestamp(data_aux + one_leap_year)
            data_str = str(datetime.datetime.fromtimestamp(data_aux + one_leap_year))
        else:
            data_contador = datetime.datetime.fromtimestamp(data_aux + one_year)
            data_str = str(datetime.datetime.fromtimestamp(data_aux + one_year))

        data_aux = data_contador.timestamp()

        # A cada iteração, calcula o juro do ano corrente
        cotacao_atual = cotacao_atual + cotacao_atual * float(juro_anual) / 100

    # Cria um dataframe com a lista de juros compostos calculados
    df_linha_teorica_juro = pd_.DataFrame(conf_)

    # Define a Data como índice do dataframe
    df_linha_teorica_juro.set_index(['Data'], inplace=True)

    # Imprime o início e o fim do dataframe
    print(df_linha_teorica_juro.head())
    print()
    print(df_linha_teorica_juro.tail())

    # Plot do gráfico
    df_linha_teorica_juro['Cotação'].plot()
    plt.legend(loc=8)
    plt.xlabel('Data')
    plt.ylabel('Linha teórica de juro')
    plt.title('Linha teórica de juro de Renda Fixa')
    plt.show()

interact(calculadora_rf, \
         cotacao_inicial=widgets.IntSlider(min=100, max=100000, step=100, value=10000), \
         juro_anual=widgets.IntSlider(min=1, max=30, step=1, value=10), \
         dia_inicio=range(1,32,1), \
         mes_inicio=[('Jan', 1), ('Fev', 2), ('Mar', 3), ('Abr', 4), ('Mai', 5), ('Jun', 6), ('Jul', 7), ('Ago', 8), ('Set', 9), ('Out', 10), ('Nov', 11), ('Dez', 12)], \
         ano_inicio=range(2020,2041,1), \
         dia_fim=range(1,32,1), \
         mes_fim=[('Jan', 1), ('Fev', 2), ('Mar', 3), ('Abr', 4), ('Mai', 5), ('Jun', 6), ('Jul', 7), ('Ago', 8), ('Set', 9), ('Out', 10), ('Nov', 11), ('Dez', 12)], \
         ano_fim=range(2040,2024,-1));
```
